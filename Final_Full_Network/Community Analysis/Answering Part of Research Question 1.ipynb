```python
import pandas as pd
import numpy as np
```

```python
nodes = pd.read_csv("nodes_with_community_connection_info.csv")
```

Target Comm Interaction Count means that you have people from many communities who retweet you

Source Comm Interaction Count means that you retweet people from many communities

```python
nodes
```

```python
comms = nodes.groupby('community')
```

```python
for name, group  in comms:
    print(name)
    ser = group['id']
    totalcount = len(ser)
    print ("Total Membership of Community: {}".format(totalcount))
    outsidespreaders = group[group['Source Comm Interaction Count'] > 0] 
    print("Total Number of people who have interacted outside of the comunity: {}".format(outsidespreaders['id'].count()))
    print("Total Number of communities: {}".format(outsidespreaders['Source Comm Interaction Count'].sum()))
    
```
