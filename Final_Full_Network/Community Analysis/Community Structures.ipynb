```python
from sklearn.metrics import classification_report, confusion_matrix
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Greens):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)

    
    thresh = cm.max() / 2.
    fmt = '.2f' if normalize else 'd'
    print(thresh)
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
        
 

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix( just_mp['political affiliation'].to_numpy(), just_mp['community'].to_numpy(), labels=["Liberal", "Conservative", "NDP", "Bloc Qubcois", "Independent", "Green"])
np.set_printoptions(precision=2)


# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=["Liberal", "Conservative", "NDP", "Bloc Qubcois", "Independent", "Green"],normalize= False,  title='Confusion matrix')
```

Go Through each possible community pairing and get a count of the edges between the two communities

```python
all_community_labels = pd.unique(communities[['community']].values.ravel())
all_community_labels
```

```python
edge_counts = {}
for c1 in all_community_labels:
    communitySourceEdges = edges_with_community[edges_with_community['source_community'] == c1]
    for c2 in all_community_labels:
        SourceAndTargetCommunity = communitySourceEdges['Metadata'][communitySourceEdges['target_community'] == c2]
        key = "{}_{}".format(c1, c2)
        if c1!=c2 and "{}_{}".format(c2, c1) in edge_counts:
            key = "{}_{}".format(c2, c1)
            edge_counts[key] = edge_counts[key] + SourceAndTargetCommunity.count()
        else: 
            edge_counts[key] = SourceAndTargetCommunity.count()
```

## Selecting the nodes that have 3 or more communities as edges

Go through each Source edge and list out the different communities its included in.

Do that for the target as well

```python
twoAndOver_sources = []
threeAndOver_sources = []

for node in edges_with_community['Source'].unique():
    # get all edges that contain source node
    nodeEdges = edges_with_community[edges_with_community['Source'] == node]
    count =  len(nodeEdges['target_community'][nodeEdges['source_community'] != nodeEdges['target_community']].unique())
    if count >= 4:
        threeAndOver_sources.append(node)
        continue
    if count >=3:
        twoAndOver_sources.append(node)
        
        
        
```
