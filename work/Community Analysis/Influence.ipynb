```python
G = nx.read_edgelist('edges_with_affiliations.csv', comments='#',
                     create_using=nx.DiGraph(), 
                     delimiter=',', 
                     nodetype=int,
                     data=[("weight", int),("target_politics",str),("source_politics",str)],
                     encoding='utf-8')
```

```python
influence = nx.algorithms.centrality.katz_centrality_numpy(G,weight="weight")
```
